```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)


import seaborn as sns
from sklearn.decomposition import PCA

pca = PCA()

pca_data = pca.fit_transform(df_scaled)
explained_var = pd.DataFrame({
    "explained_variance": pca.explained_variance_,
    "explained_variance_ratio": pca.explained_variance_ratio_,
    "explained_variance_ratio_cumsum": pca.explained_variance_ratio_.cumsum()
    
}, index=range(1, len(pca.explained_variance_) + 1))

explained_var
plt.figure(figsize=(12,7))
plt.plot(range(1, len(pca.explained_variance_) + 1), explained_var['explained_variance_ratio_cumsum']*100)
plt.scatter(range(1, len(pca.explained_variance_) + 1), explained_var['explained_variance_ratio_cumsum']*100)
plt.xlabel('Фактор')
plt.ylabel('Суммарный % дисперсии')
plt.xticks(range(1, len(pca.explained_variance_) + 1))
plt.title('График объясненной дисперсии')
plt.show()
```

```python
from sklearn.decomposition import FactorAnalysis

n = 6

faa = FactorAnalysis(n_components=n)
faa.fit_transform(df_scaled)
df_loadings = pd.DataFrame(faa.components_.T, 
                          index=df_scaled.columns,
                          columns=[f'Factor{i+1}' for i in range(n)])
plt.figure(figsize=(20,20))
sns.heatmap(df_loadings, annot=True, cmap='coolwarm')
df_loadings
```

```python
faa = FactorAnalysis(n_components=n, rotation='varimax')
faa.fit_transform(df_scaled)
df_loadings = pd.DataFrame(faa.components_.T, 
                          index=df_scaled.columns,
                          columns=[f'Factor{i+1}' for i in range(n)])
plt.figure(figsize=(20,20))
sns.heatmap(df_loadings, annot=True, cmap='coolwarm')
df_loadings
```
